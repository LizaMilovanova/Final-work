```python
alt.renderers.enable('default')

df = data.wheat()

bars = alt.Chart(df).mark_bar().encode(
    x='wheat:Q',
    y='year:O'
)

text = alt.Chart(df).mark_text(
    align='left',
    dx=3
).encode(
    x='wheat:Q',
    y='year:O',
    text='wheat:Q'
)

bars + text
```

```python
iowa_data = data.iowa_electricity()

chart2 = alt.Chart(iowa_data).mark_area(opacity=0.7).encode(
    x='year:T',
    y='net_generation:Q',
    color='source:N'
).properties(title='Виробництво електроенергії в Айові за джерелами')

chart2.display()
```

```python
df_iris = px.data.iris()

fig = px.scatter(
    df_iris,
    x='sepal_width',
    y='sepal_length',
    color='species',
    size='petal_length',
    hover_data=['petal_width'],
    title='Аналіз датасету Ірисів'
)

fig.show()
```
