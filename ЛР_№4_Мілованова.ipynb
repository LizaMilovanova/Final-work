```python
df = df.fillna(df.mean(numeric_only=True), axis=0)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('/', '_').str.replace('__', '_').str.strip()
df.columns
```

```python
df.tail()
```

```python
df['month'] = pd.to_datetime(df['month'], format='%m/%d/%Y %I:%M:%S %p')
df_2018 = df[df['month'].dt.year == 2018]

min_visitors = df_2018.iloc[:,1].min()
max_visitors = df_2018.iloc[:,1].max()
result = pd.DataFrame({'min': min_visitors, 'max': max_visitors})
result
```

```python
average_visitors = df.iloc[:, 1:].mean()
average_visitors
```

```python
df_2015 = df[df['month'].dt.year == 2015].copy()
df_2015['total_visitors'] = df_2015.iloc[:, 1:].sum(axis=1)
min_row = df_2015.loc[df_2015['total_visitors'].idxmin()]
max_row = df_2015.loc[df_2015['total_visitors'].idxmax()]

min_month = min_row['month'].strftime('%B')
max_month = max_row['month'].strftime('%B')

print(f'Найнижча відвідуваність у {min_month}')
print(f'Найвища відвідуваність у {max_month}')
```

```python
df_2015.head()
```

```python
df_2018 = df[df['month'].dt.year == 2018].copy()

df_2018['month_num'] = df_2018['month'].dt.month

summer = df_2018[df_2018['month_num'].isin([6, 7, 8])]['avila_adobe'].sum()

winter = df_2018[df_2018['month_num'].isin([1, 2, 12])]['avila_adobe'].sum()

print(f" Літом відвідали: {summer} людей")
print(f" Зимою відвідали: {winter} людей")
```

```python
mtx = df.corr(numeric_only=True).abs()
fig,ax = plt.subplots(figsize=(8,8))

sns.heatmap(
    mtx,
    cmap='crest',
    annot=True,
    fmt='.2f',
    linewidths=0.5,
    mask=np.triu(np.ones_like(mtx,dtype=bool)),
    square=True,
    cbar=False,
    ax=ax
)
plt.title("Матриця попарної кореляції ознак", fontsize=14)
plt.tight_layout()
plt.show
```

```python
sns.set_theme(style='darkgrid')
sns.lineplot(data=df, x='month', y='avila_adobe')
plt.title('Avila Adobe: Відвідуваність за часом')
plt.show()
```

```python
df['year'] = df['month'].dt.year
sns.boxplot(data=df,  x='year', y='chinese_american_museum')
plt.title('Chinese American Museum: Розповіл по роках')
plt.show()
```

```python
sns.histplot(df['gateway_to_nature_center'], bins=30, kde=True)
plt.title('Розподіл відвідуваності Gateway to Nature Center')
plt.show()
```

```python
avg_visitors = df.drop(columns=['month','year']).mean().sort_values(ascending=False)
sns.barplot(x=avg_visitors.values, y=avg_visitors.index)
plt.title('Середня кількість відвідувачів по музеях')
plt.xlabel('Середнє значення')
plt.ylabel('Музей')
plt.show()
```

```python
task_2 = pd.read_csv('penguins.csv')
task_2
```

```python
task_2.info()
```

```python
task_2 = task_2.fillna(task_2.mean(numeric_only=True), axis=0)
```

```python
task_2.isna().sum()
```

```python
# 1. KDE-графік розподілу маси по видам
plt.figure()
sns.kdeplot(data=task_2, x='body_mass_g', hue='species', fill=True)
plt.title('Density of Body Mass by Species')
plt.show()
```

```python
# 2. Залежність довжини дзьоба від глибини
plt.figure()
sns.scatterplot(data=task_2, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
# 3. Violin plot
plt.figure()
sns.violinplot(data=task_2, x='species', y='body_mass_g', hue='sex', split=True)
plt.title('Body Mass Distribution by Species and Sex')
plt.show()
```

```python
# 4. Довжина плавника по островах
plt.figure()
sns.barplot(data=task_2, x='island', y='flipper_length_mm')
plt.title('Flipper Length by Island')
plt.show()
```

```python
# 5. Heatmap кореляцій
plt.figure()
corr = task_2.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
