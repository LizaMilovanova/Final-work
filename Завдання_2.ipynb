```python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
print("Нові назви стовпців:", df.columns)
```

```python
# Гістограма віку
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], kde=True, color='skyblue')
plt.title('Розподіл віку клієнтів')
plt.show()
```

```python
# Порівняння розподілу покупок
plt.figure(figsize=(12, 6))
sns.countplot(y=df['item_purchased'], order=df['item_purchased'].value_counts().index, palette='viridis')
plt.title('Частота покупок різних товарів')
plt.show()
```

```python
# Залежність від віку та сумою покупок
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='purchase_amount_(usd)', alpha=0.6)
plt.title('Залежність суми покупки від віку')
plt.show()
```

```python
# Графік розподілу оцінок товарів для оцінки якості
plt.figure(figsize=(10, 5))
sns.kdeplot(df['review_rating'], fill=True, color='olive')
plt.title('Розподіл оцінок товарів (якість)')
plt.show()
```

```python
# Категоріальна діаграма для статі
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gender', palette='pastel')
plt.title('Розподіл покупок за статтю')
plt.show()
```

```python
# Залежність між рейтингом і сумою покупок
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='review_rating', y='purchase_amount_(usd)', hue='gender')
plt.title('Звязок рейтингу та суми покупки')
plt.show()
```
