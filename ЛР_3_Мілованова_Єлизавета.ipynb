1. Побудувати графіки для тригонометричних функцій, використати різні стилі та маркери

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 180)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, y1, linestyle='-', marker='o', markevery=20, label='sin(x)')
ax.plot(x, y2, linestyle='--', marker='s', markevery=20, label='cos(x)')

ax.set_title("Дві лінії")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

2. Побудуйте точкову діаграму, де колір точки залежить від третьої змінної, додайте шкалу кольорів.

```python
np.random.seed(1)
x = np.random.randn(180)
y = 1.2 * x + np.random.randn(180) / 2

c = np.hypot(x, y)  # “інтенсивність” для кольору

fig, ax = plt.subplots(figsize=(6, 3))
sc = ax.scatter(x, y, c=c, s=50, edgecolor='k')
ax.set_title("Scatter: колір за величиною")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.colorbar(sc, ax=ax, label="шкала кольору")
ax.grid(True, alpha=0.3)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Створення сітки значень X та Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Визначення функції Z = sin(sqrt(X^2 + Y^2))
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Побудова 3D-поверхні
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Побудова поверхні
surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Додавання колірної шкали та підписів
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5, label='Z')
ax.set_title(r'3D-поверхня $Z = \sin(\sqrt{X^2 + Y^2})$')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

**Dashboard з GridSpec**: 4 графіки + спільний заголовок

Умова: Створіть одну фігуру-«дашборд» через GridSpec:

зверху широкий графік тренду (line),

знизу зліва гістограма,

знизу справа scatter,

праворуч вузька панель-«індикатор» (bar).

Додайте спільний заголовок фігури, сітку та охайні підписи.

```python
np.random.seed(0)
t = np.linspace(0, 20, 300)
trend = 0.15*t + np.sin(t) + np.random.normal(0, 0.3, len(t))

x = np.random.randn(400)
y = 0.8*x + np.random.randn(400)/2

# Створення фігури
from matplotlib.gridspec import GridSpec
fig = plt.figure(figsize=(12, 8))
# Визначаємо сітку: 2 рядки, 3 стовпці.
# Третій стовпець зробимо вужчим за допомогою width_ratios
gs = GridSpec(2, 3, figure=fig, width_ratios=[1, 1, 0.4], wspace=0.3, hspace=0.3)

# 1. Зверху широкий графік тренду (Line plot)
ax1 = fig.add_subplot(gs[0, :2])
ax1.plot(t, trend, color='tab:blue', label='Trend line')
ax1.set_title('Графік тренду')
ax1.set_xlabel('Час')
ax1.set_ylabel('Значення')
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend()

# 2. Знизу зліва гістограма (Histogram)
ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(x, bins=30, color='tab:green', alpha=0.7, edgecolor='black')
ax2.set_title('Гістограма розподілу X')
ax2.set_xlabel('Значення X')
ax2.set_ylabel('Частота')
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# 3. Знизу справа scatter (Scatter plot)
ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(x, y, s=10, alpha=0.5, color='tab:red')
ax3.set_title('Кореляція X та Y')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.grid(True, linestyle='--', alpha=0.7)

# 4. Праворуч вузька панель-«індикатор» (Bar plot)
ax4 = fig.add_subplot(gs[:, 2])
categories = ['A', 'B', 'C', 'D']
values = [15, 24, 18, 30]
ax4.bar(categories, values, color='tab:orange', alpha=0.8)
ax4.set_title('Індикатори')
ax4.set_ylabel('Показник')
ax4.grid(axis='y', linestyle='--', alpha=0.7)

# Загальний заголовок та підписи
fig.suptitle('Аналітичний дашборд даних', fontsize=16, fontweight='bold')
```

**Завдання 18.**

Зробіть анімацію, де точка «їде» по кривій sin(x).
Потрібно: лінія графіка + рухома точка + текст з координатами точки.

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)
ax.plot(x, y, color='blue', alpha=0.3)
ax.grid(True, linestyle='--')

point, = ax.plot([], [], 'ro', markersize=10)
text = ax.text(0, 0, '', fontsize=12)

# Функції анімації
def init():
    point.set_data([], [])
    text.set_text('')
    return point, text

def update(frame):
    curr_x = x[frame]
    curr_y = y[frame]
    point.set_data([curr_x], [curr_y])
    text.set_position((curr_x + 0.1, curr_y + 0.1))
    text.set_text(f'x={curr_x:.2f}, y={curr_y:.2f}')
    return point, text

# Створення анімації
ani = FuncAnimation(fig, update, frames=len(x), init_func=init, blit=True, interval=50)

plt.close()
HTML(ani.to_jshtml())
```
