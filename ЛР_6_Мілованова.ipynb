```python
#Сніжинка Коха
from ColabTurtlePlus.Turtle import *

clearscreen()
def koch_curve(t, iterations, length):
    if iterations == 0:
        t.forward(length)
    else:
        for angle in [60, -120, 60, 0]:
            koch_curve(t, iterations - 1, length / 3)
            t.left(angle)

def draw_snowflake(iterations, length):
    setup(500, 500)
    t = Turtle()
    t.speed(10)

    t.penup()
    t.goto(-150, 100)
    t.pendown()

    for _ in range(3):
        koch_curve(t, iterations, length)
        t.right(120)

draw_snowflake(3, 300)
```

```python
#Дерево Піфагора
from ColabTurtlePlus.Turtle import *

clearscreen()
setup(500, 500)

def draw_tree(branch_len, t):
    if branch_len > 5:
        t.forward(branch_len)
        t.right(20)
        draw_tree(branch_len - 15, t)
        t.left(40)
        draw_tree(branch_len - 15, t)
        t.right(20)
        t.backward(branch_len)

def main():
    t = Turtle()
    t.speed(13)
    t.left(90)
    t.penup()
    t.goto(0, -200)
    t.pendown()
    t.color("green")

    draw_tree(75, t)
    t.hideturtle()

main()
```

```python
#Трикутник Серпінського
from ColabTurtlePlus.Turtle import *

clearscreen()
setup(500, 500)

def sierpinski(t, order, size):
    if order == 0:
        for _ in range(3):
            t.forward(size)
            t.left(120)
    else:
        sierpinski(t, order - 1, size / 2)
        t.forward(size / 2)
        sierpinski(t, order - 1, size / 2)
        t.backward(size / 2)
        t.left(60)
        t.forward(size / 2)
        t.right(60)
        sierpinski(t, order - 1, size / 2)
        t.left(60)
        t.backward(size / 2)
        t.right(60)

def main():
    t = Turtle()
    t.speed(13)

    t.penup()
    t.goto(-150, -100)
    t.pendown()
    t.color("orange")

    sierpinski(t, 4, 300)
    t.hideturtle()

main()
```
